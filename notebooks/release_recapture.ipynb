```julia
include("../src/ColdAtoms.jl")
using .ColdAtoms 
```

```julia
using PyPlot
```

```julia
U0, w0, λ, M2 = 1000.0, 1.1, 0.852, 1.3;
z0 = ColdAtoms.w0_to_z0(w0, λ, M2);
m, T = 87.0, 50.0;
trap_params = [U0, w0, z0];
atom_params = [m, T];
```

```julia
tspan = [0.0:1.0:40.0;];
N = 10000 # Number of samples 

probs = ColdAtoms.release_recapture(tspan, trap_params, atom_params, N; skip=1000, harmonic=false, eps=1e-3)[1];
plot(tspan, probs)
```
