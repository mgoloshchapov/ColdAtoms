```julia
using PyPlot
using CSV, DataFrames
```

```julia
filename_temperature = "../data/2023_12_5_N54_RPow600k_rez183_rydberg_r&r_deep.dat"



df_temperature = DataFrame(CSV.File(filename_temperature, delim=" ", header=26, ignorerepeated=true, types=Float64))
x = df_temperature[1:end, 1];
y = df_temperature[1:end, 6];
```

```julia
fontsize=16
figure(figsize=(7,4), dpi=200)
plot(x, y, color="red")
scatter(x, y, color="red")

xlim(0.0, 30.0)
ylim(0.0, 1.0)

text(11, 0.6, "\$ t_{recap} = 4.2\$ μs", fontsize=fontsize-2)
text(12, 0.5, "\$ \\varepsilon' = 2-3 \\% \$", fontsize=fontsize-2)
xlabel("Time, μs", fontsize=fontsize)
ylabel("Recapture probability", fontsize=fontsize)

tight_layout()
savefig("/Users/goloshch/Desktop/Materials/Workshops:Conferences/2023/ФУХА/presentation/images/rr_antitrap_eng.png", dpi=200)
gcf()
```
