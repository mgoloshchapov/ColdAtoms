```julia
using PyPlot
include("../src/ColdAtoms.jl");
using .ColdAtoms
using CSV, DataFrames
```

```julia
filename1 = "../data/temperature/04_02_2024_N54_trap805nm_temperature.dat";
filename2 = "../data/temperature/05_02_2024_N54_TiSa_trap805nm_temperature.dat";

df1 = DataFrame(CSV.File(filename1, delim=" ", header=20, ignorerepeated=true, types=Float64))
df2 = DataFrame(CSV.File(filename2, delim=" ", header=20, ignorerepeated=true, types=Float64))

x1 = df1[1:end, 1];
y1 = df1[1:end, 2];

x2 = df2[1:end, 1];
y2 = df2[1:end, 2];
```

```julia
fontsize = 14;
linewidth = 2.0;

T0 = 60.0;

atom_params_temp0 = [86.9091835, T0];
trap_params_temp = [700, 1.0, 4.2]

N = 20000;
prob0, _ = release_recapture(x1, trap_params_temp, atom_params_temp0, N; freq=50, skip=5000)

figure(figsize=(8,4), dpi=120)
scatter(x1, y1 ./ maximum(y1), color="red", s=10, alpha=0.5, zorder=2, label="Данные")
plot(x1, prob0, color="black", linewidth=linewidth, linestyle="--", label="Монте-Карло, T\$=60 \$ мкК")
xlabel("Время выключения, мкс", fontsize=fontsize)
ylabel("Вероятность остаться", fontsize=fontsize)
xticks(fontsize=fontsize-2)
yticks(fontsize=fontsize-2)
xlim(0.0, maximum(x1))
ylim(0.0, 1.0)
grid(alpha=0.2)
legend(fontsize=fontsize-2)

text(20.0, 0.5, "\$ U_0 = 700 \$ мкК", fontsize=fontsize)
text(20.0, 0.4, "\$ w_0 = 1.0 \$ мкм", fontsize=fontsize)
title("04_02_2024_N54_trap805nm_temperature.dat", fontsize=fontsize)

tight_layout()

savefig("../data/temperature/temperature_805nm.pdf", dpi=300)
# savefig("/Users/goloshch/Desktop/Materials/QOTLabs/Деловое/Отчет_23/images/temperature.png", dpi=300)
# savefig("/Users/goloshch/Desktop/Materials/QOTLabs/Деловое/Отчет_23/images/temperature.pdf", dpi=300)
# savefig("/Users/goloshch/Desktop/Materials/QOTLabs/Деловое/Отчет_23/images/temperature.svg", dpi=300)
gcf()
```
